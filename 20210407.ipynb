# 1.カスタム層

単純なカスタム層を実装してみる

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

class CustomLayer(layers.Layer):
    def __init__(self, width, **kwargs):
        super().__init__(**kwargs)
        self.width = width

    def build(self, batch_input_shape):
        self.kernel = self.add_weight(shape=[batch_input_shape[-1],self.width],
            initializer='glorot_normal')
        super().build(batch_input_shape)

    def call(self, X):
        return X @ self.kernel
```

これを使った一層モデル

```python
inputs = layers.Input(shape=(3,))
x = CustomLayer(5)(inputs)
model = keras.Model(inputs=inputs, outputs=x)

model.summary()
```

で次のような人工データをフィッティングしてみる：

```python
X = np.random.randn(3*100).reshape((100,3))
W = np.random.randn(3*5).reshape((3,5))
Y =  X@W
```

もちろんうまくいけばこのデータを作るのに使用した行列Wとモデルの重み行列は一致するはずである。

学習させてみよう：

```python
model.compile(optimizer='adam', loss='mse')
model.fit(X,Y,epochs=500)
```

学習済みのカスタム層の重みを見てみると、その値は

```python
model.layers[1].kernel
```

テンソルではなく、数値だけ見たければ

```python
model.layers[1].kernel.numpy()
```

データの作成に使った行列と比べてみると

```python
W
```

# 2. サブクラス化API

次に基底Modelクラスを継承してカスタムモデルを作ってみよう。と言っても、簡単なことをしているうちは今までのfunctional API的なモデルを一つにまとめたような感じしか受けないでしょう。

```python
class SimpleModel(keras.Model):
    def __init__(self, width=5, **kwargs):
        super(SimpleModel, self).__init__(**kwargs)
        self.hidden = CustomLayer(width)

    def call(self, inputs):
        outputs = self.hidden(inputs)
        return outputs
```

大きなポイントは、入力層を用意する必要がない、あらわに入力サイズを指定する必要がないという点です。そのためにモデル作成時に自動でshapeのチェックをすることができません。また、同時にモデルの構造がcall()メソッドの中で定義されているので、構造が外から見えなくなってしまいます。このような不自由な点があるため、普段はfunctional APIを使っておくのが良いでしょう（今までのようにモデルを作るのでは事足りない時にSubclassing APIを使いましょう。つまり例えばモデルの構造が入力によって条件分岐して変わるようなダイナミックなモデルです。このようなモデルは、Subclassing APIのcall()メソッドの柔軟性を活用すると綺麗に実装・利用できます）。

いずれにせよ、モデルのインスタンスは普通に作成でき

```python
model = SimpleModel(5)
```

学習もいつも通りです：

```python
model.compile(optimizer='adam', loss='mse')
model.fit(X,Y,epochs=500)
```

カスタムモデルクラスには入力層はないので、次の引数が変わることに注意

```python
model.layers[0].kernel
```

当然、学習結果はほとんど変わらない。
